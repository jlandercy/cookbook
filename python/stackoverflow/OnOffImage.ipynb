```python
import itertools
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

def generate(N=2000, threshold=.3, x=4, y=3):
    
    alpha = np.linspace(-10, 10, N)
    beta = np.linspace(-10, 10, N)
    A, B = np.meshgrid(alpha, beta)

    D = np.sin(B * np.pi / 180.) * x + np.sin(A * np.pi / 180.) * np.cos(B * np.pi * 180.) * y
    S = D > threshold
    
    return S

S = generate(N=2000)

fig, axe = plt.subplots()
axe.imshow(S, cmap=plt.cm.gray)
```

```python
fig, axes = plt.subplots(2, 3)
for axe, N in zip(itertools.chain(*axes), [100, 200, 300, 500, 1000, 2000]):
    S = generate(N=N)
    axe.imshow(S, cmap=plt.cm.gray)
    axe.set_title("N = %d" % N)
```
