```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

https://stackoverflow.com/questions/78233708/getting-a-weird-error-while-using-solve-ivp-from-scipy-integrate-after-qr-from-n

```python
def model(t, x, s, r, b):
    xdot = np.array([
        s * (x[1] - x[0]),
        x[0] * (r - x[2]) - x[1],
        x[0] * x[1] - b * x[2]
    ])
    return xdot
```

```python
s=10; r=28; b = 8/3
x0 = np.random.randn(3)
T = 20
t = np.linspace(0, T, 2000)
```

```python
solution = integrate.solve_ivp(model, [0, T], x0, args=(s, r, b), t_eval=t, dense_output=True, atol=1e-9, rtol=1e-6)
```
