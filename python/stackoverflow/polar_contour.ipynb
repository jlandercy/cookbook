```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0, 2*np.pi, 30)
rho = np.linspace(0.4, 1.0, 20)
t, r = np.meshgrid(theta, rho)
```

```python
f = r*np.sin(t)
```

```python
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.contourf(t, r, f, cmap='jet')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

pi = math.pi
ang = pi/16.0
num = 8
  
rr = np.arange(0.4, 1.0, 0.05)
thetaa = np.arange(0,   2*pi + 0.1,   ang)

values = []
    
for i in range(rr.size):
      
    for j in range(thetaa.size):
        
        dum_v = i * j
        values.append(dum_v)    
    
r, theta = np.meshgrid(rr, thetaa)

values = np.array(values)
values = values.reshape(r.shape)


fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.axis([0, 2*pi, 0.0, 0.8])
ax.contourf(theta,r,values)
```
