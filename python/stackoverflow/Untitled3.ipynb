```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate
```

 - https://docs.scipy.org/doc/scipy/reference/interpolate.html

```python
x = np.arange(0, 11)
y = np.arange(0, 11)
X, Y = np.meshgrid(x, y)
Z = np.ones(X.shape)
Z[5,5] = 9
```

```python
fig, axe = plt.subplots()
axe.imshow(Z, origin="lower", cmap="jet")
```

```python
fig, axe = plt.subplots()
axe.imshow(Z, origin="lower", cmap="jet", interpolation='bicubic')
```

```python
interpolator = interpolate.interp2d(x, y, Z.ravel(), kind="linear")
```

```python
interpolator(x, y)
```

```python
xlin = np.linspace(0, 10, 101)
ylin = np.linspace(0, 10, 101)
```

```python
zhat = interpolator(xlin, ylin)
zhat
```

```python
fig, axe = plt.subplots()
axe.imshow(zhat, origin="lower", cmap="jet")
```

```python
Xlin, Ylin = np.meshgrid(xlin, ylin)
smooth = interpolate.NearestNDInterpolator([r for r in zip(X.ravel(), Y.ravel())], Z.ravel(), rescale=True)
Zs = smooth([r for r in zip(Xlin.ravel(), Ylin.ravel())]).reshape(Xlin.shape)
```

```python
fig, axe = plt.subplots()
axe.imshow(Zs, origin="lower", cmap="jet", interpolation='bicubic')
```

```python
x = np.arange(0, 11)
y = np.arange(0, 11)
X, Y = np.meshgrid(x, y)
Z = np.ones(X.shape)
Z[5,5] = 9
```

```python
x = X.ravel()
y = Y.ravel()
z = Z.ravel()
```

```python
z.shape
```

```python
resolution = 11
xlin = np.linspace(x.min(), x.max(), resolution)
ylin = np.linspace(y.min(), y.max(), resolution)
Xlin, Ylin = np.meshgrid(xlin, ylin)
```

```python
interpolant = interpolate.NearestNDInterpolator([r for r in zip(x, y)], z)
```

```python
Zhat = interpolant(Xlin.ravel(), Ylin.ravel()).reshape(Xlin.shape)
```

```python
fig, axe = plt.subplots()
axe.imshow(Zhat, origin="lower", cmap="jet", interpolation='bicubic')
```
