```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
rate, data = wavfile.read('./data/aaaah.wav') # See source
rate, data
```

```python
_ = plt.specgram(data[:,0])
```

```python
d = []
for k in range(0, data.shape[0]+1, 256):
    x = np.fft.fft(data[k:k+256,0])
    d.append(np.log10(np.real(x[128:256]*np.conj(x[128:256]))))
d.pop(-1)
```

```python
d = np.rot90(np.array(d))
```

```python
plt.imshow(d)
```
