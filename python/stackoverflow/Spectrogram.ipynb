```python
import numpy as np
from scipy.io import wavfile
from scipy import fft
import matplotlib.pyplot as plt
```

 - https://stackoverflow.com/questions/63640027/spectrogram-in-python-using-numpy/63640628#63640628

```python
rate, data = wavfile.read('./data/aaaah.wav') # See source
rate, data.shape
```

```python
N = 256
d = []
for k in range(0, data.shape[0]+1, N):
    x = fft.fftshift(fft.fft(data[k:k+N,0], n=N))[N//2:N]
    Pxx = 10*np.log10(np.real(x*np.conj(x)))
    d.append(Pxx)
d = np.array(d)
```
