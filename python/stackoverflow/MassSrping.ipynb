```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate 
```

```python
def system(t, x, k, m):
    return np.array([
        x[1],
        -k/m * x[0]
    ])
```

```python
t = np.linspace(0, 10, 200)
```

```python
s = integrate.solve_ivp(system, [t.min(), t.max()], [1, 1], t_eval=t, args=(1, 1))
```

```python
fig, axe = plt.subplots()
axe.plot(s.t, s.y.T)
```
