```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
depth = pd.read_csv("Bot3Dive2At1m_Depth.csv", header=None, names=["t", "z"]).set_index("t")
temp = pd.read_csv("Bot3Dive1At1m_Temperature.csv", header=None, names=["t", "T"]).set_index("t")
```

```python
data = depth.merge(temp, left_index=True, right_index=True)
```

```python
data["z"].plot()
```

```python
data["T"].plot()
```

```python
fig, axe = plt.subplots()
axe.scatter(data["z"], data["T"])
```
