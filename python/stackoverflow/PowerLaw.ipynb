```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def model(x, alpha):
    return (alpha - 1.) * np.power(x[:, 0], -alpha)
```

```python
np.random.seed(12345)
X = np.linspace(1, 50, 30).reshape(-1, 1)
y = model(X, 2.5)
s = 0.05 * np.abs(y)
yn = y + s * np.random.normal(size=y.shape) 
```

```python
popt, pcov = optimize.curve_fit(model, X, yn, sigma=s, absolute_sigma=True)
popt, pcov
```

```python
popt, pcov = optimize.curve_fit(model, X, yn, sigma=s, p0=(3.,), absolute_sigma=True)
popt, pcov
```

```python
fig, axe = plt.subplots()
axe.loglog(X, yn)
```

```python
Xlin = np.linspace(1, 50, 200).reshape(-1, 1)
```

```python
alphas = np.linspace(1., 3.0, 13)
```

```python
fig, axe = plt.subplots()
for alpha in alphas:
    axe.loglog(Xlin, model(Xlin, alpha), color="gray", linestyle="--")
axe.scatter(X, yn)
axe.grid(which="both")
```
