```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def model1(t, x, gamma, omega):
    return np.array([
        x[1],
        - gamma * x[1] - omega ** 2 * np.sin(x[0])
    ])
```

```python
def model2(t, x, gamma, omega):
    return np.array([
        x[1],
        - gamma * x[1] - omega ** 2 * x[0]
    ])
```

```python
t = np.linspace(0, 40, 5000)
```

```python
p = (0.025, 3.12)
```

```python
a0 = 5. / 180. * np.pi
a0
```

```python
sol1 = integrate.solve_ivp(model1, [t.min(), t.max()], [a0, 0.], args=p, t_eval=t)
sol2 = integrate.solve_ivp(model2, [t.min(), t.max()], [a0, 0.], args=p, t_eval=t)
```

```python
fig, axe = plt.subplots()
axe.plot(sol1.t, sol1.y[0])
axe.plot(sol2.t, sol2.y[0])
axe.grid()
```
