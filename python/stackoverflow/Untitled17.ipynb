```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, fft
```

```python
x = stats.uniform().rvs(size=10_000)
x = x - x.mean()
```

```python
X = fft.fft(x)
```

```python
fig, axe = plt.subplots()
axe.plot(X * np.conj(X))
```
