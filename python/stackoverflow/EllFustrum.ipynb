```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
u = np.linspace(0, 2 * np.pi, 200)
v = np.linspace(0, 1 * np.pi, 200)
U, V = np.meshgrid(u, v)
```

```python
X = np.cos(U) * np.sin(V)
Y = np.sin(U) * np.sin(V)
Z = np.cos(V)
```

```python
fig, axe = plt.subplots(subplot_kw={"projection": "3d"})
axe.plot_surface(X, Y, Z, alpha=0.5)
```
