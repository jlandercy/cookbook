```python
xmin, xmax = -7, 7
x = y = np.linspace(xmin, xmax, 100)
X, Y = np.meshgrid(x, y)
```

```python
xh = yh = np.linspace(xmin, xmax, 400)
Xh, Yh = np.meshgrid(xh, yh)
```

```python
def z(x, y):
    return np.power(np.sin(x), 3) + np.power(np.cos(y), 3)
```

def z(x, y):
    return x*y

```python
Z = z(X, Y)
Zh = z(Xh, Yh)
```

```python
dZ = np.gradient(Z, x, y)
M = np.hypot(*dZ)
```

```python
fig, axe = plt.subplots(figsize=(12, 12))
axe.contour(Xh, Yh, Zh, 30, cmap="jet", linewidths=0.75)
axe.quiver(X, Y, dZ[1], dZ[0], M, cmap="jet", units='x', pivot='tail', width=0.03, scale=5)
axe.set_aspect("equal")
axe.grid()
```

```python
fig = plt.figure()
axe = plt.axes(projection='3d')
axe.plot_surface(X,Y,Z)
```
