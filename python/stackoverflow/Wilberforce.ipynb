```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft, signal, optimize
```

```python
def model(t, A, omega1, omega2, phi):
    return A * (np.sin(omega1 * t + phi) + np.sin(omega2 * t + phi))
```

```python
dt = 40
N = 400
```

```python
t = np.linspace(0, dt, N)
p = (10, 2 * np.pi, 1.05 * 2 *np.pi, 0.1)
x = model(t, *p)
```

```python
fs = N / dt
```

```python
f, P = signal.welch(x, fs)
```

```python
fig, axe = plt.subplots()
axe.plot(f, P)
axe.grid()
```

```python
fig, axe = plt.subplots()
axe.plot(t, x)
axe.grid()
```
