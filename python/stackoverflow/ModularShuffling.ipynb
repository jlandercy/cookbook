```python
class Shuffler:
    
    @staticmethod
    def new(n):
        return np.array(["B"]*n + ["R"]*n)
    
    @staticmethod
    def shuffle(x):
        return x.reshape((2, -1)).T.reshape((2, -1)).flatten()
    
    @staticmethod
    def identity(n):
        return np.arange(2*n) + 1
    
    @staticmethod
    def permutation(n):
        return Shuffler.shuffle(Shuffler.identity(2*n))
    
    @staticmethod
    def generate(x=None, n=1):
        
        if x is None:
            x = Shuffler.new(n)
        else:
            n = x.size // 2
        assert x.size == 2*n
            
        x0 = x.copy()
        yield x0
        
        while True:
            x = Shuffler.shuffle(x)
            yield x
            
            if np.all(x == x0):
                break
    
    @staticmethod
    def length(x):
        n = 0
        for x in Shuffler.generate(x):
            n += 1
        return n - 1
    
    @staticmethod
    def screen(n):
        t = np.arange(n) + 1
        c = np.full(t.shape, 0)
        for i, m in enumerate(t):
            x = Shuffler.new(m)
            c[i] = Shuffler.length(x)
        return t, c
```
