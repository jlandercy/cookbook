```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
from matplotlib.pyplot import Figure
```

```python
df = pd.DataFrame({"key": ["A"], "lon": [3.708095], "lat": [43.404495]})
```

```python
point = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs=4326).to_crs(crs="ESRI:102014")
circle1 = point.geometry.buffer(1500)
circle2 = point.geometry.buffer(3000)
```

```python
axe = point.plot(color="red")
circle2.plot(alpha=0.15, color="#00b1aa", ax=axe)
circle2.boundary.plot(color="#00b1aa", ax=axe)
circle1.plot(alpha=0.35, color="#e26056", ax=axe)
circle1.boundary.plot(color="#e26056", ax=axe)
ctx.add_basemap(axe, zoom="auto", crs=point.crs, attribution_size=5, source=ctx.providers.OpenStreetMap.Mapnik)
axe.set_axis_off()
axe.figure.tight_layout()
```
