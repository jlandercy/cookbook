```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
```

```python
df = pd.DataFrame({"key": ["A"], "lon": [4.00], "lat": [51.00]})
```

```python
point = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs=4326).to_crs(crs=31370)
point = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs=4326).to_crs(crs="ESRI:102014")
#point = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs=4326).to_crs(crs=6347)
circle = point.geometry.buffer(1500)
```

```python
axe = point.plot()
circle.plot(alpha=0.3, ax=axe)
ctx.add_basemap(axe, zoom=14, crs=point.crs)
axe.set_axis_off()
```
