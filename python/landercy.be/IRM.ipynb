```python
import re
import requests
import urllib.parse
import pandas as pd
```

```python
stop = pd.Timestamp.now()
start = stop.round('1d') - pd.Timedelta('7d')
```

```python
url = "https://opendata.meteo.be/service/ows?service=WFS&version=2.0.0&request=GetFeature&typenames=synop:synop_data&outputformat=csv&CQL_FILTER=((BBOX(the_geom,2.887306,49.620358,6.073438,51.347375,%20%27EPSG:4326%27))%20AND%20(timestamp%20%3E=%20%27{start:%Y-%m-%d}%2000:00:00.000%27%20AND%20timestamp%20%3C=%20%27{stop:%Y-%m-%d}%2000:00:00.000%27))&sortby=timestamp".format(start=start, stop=stop)
```

```python
data = pd.read_csv(url)
```

```python
data["timestamp"] = pd.to_datetime(data["timestamp"])
```
