## Random Variables

Distributions are represented by random variables objects either by a generator or a frozen instance of class exposing [several useful methods][1].

[1]: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.html#scipy.stats.rv_continuous

### Uniform distribution

Lets start with a uniform distribution:

```python
X1 = stats.uniform()
```

If we want draw a sample of size 100 from this distribution, we use `rvs` method (`random_state` setting allow to setup the peudo-random generator seed in order to keep sample identical among runs):

```python
x1 = X1.rvs(1000, random_state=1)
x1g = np.linspace(-0.1, 1.1, 1000)
```

Probability densitity and cumulative functions are available respectivilty by `pdf` and `cdf` methods: 

```python
def plotDistribution(rv, sample, grid):
    fig, axe = plt.subplots(1, 3, sharey=True, gridspec_kw={'width_ratios': [5,1,1], 'wspace': 0.05})
    axe[0].plot(sample, '.')
    for i in range(3):
        axe[i].axhline(rv.mean(), color='red')
        axe[i].axhline(rv.mean() + rv.std(), color='orange')
        axe[i].axhline(rv.mean() - rv.std(), color='orange')
    axe[1].plot(rv.pdf(grid), grid)
    axe[2].plot(rv.cdf(grid), grid)
    axe[0].set_xlabel('Sample Index, $i$')
    axe[0].set_ylabel('Random Variable, $x_i$')
    axe[1].set_xlabel('PDF, $f(x)$')
    axe[2].set_xlabel('CDF, $F(x)$')
    return axe[0]
plotDistribution(X1, x1, x1g)
```

### Normal Distribution

```python
X2 = stats.norm()
x2 = X2.rvs(1000, random_state=1)
x2g = np.linspace(-5, 5, 1000)
plotDistribution(X2, x2, x2g)
```

### Cauchy Distribution

```python
X5 = stats.cauchy()
x5 = X5.rvs(1000, random_state=1)
x5g = np.linspace(-5, 5, 1000)
axe = plotDistribution(X5, x5, x5g)
axe.set_ylim([-5,5])
```

```python
axe = plotDistribution(X5, x5, np.linspace(-3000, 3000, 10000))
```

### Log Normal Distribution

```python
X3 = stats.lognorm(0.5)
x3 = X3.rvs(1000, random_state=1)
x3g = np.linspace(0, 10, 1000)
plotDistribution(X3, x3, x3g)
```

```python
X4 = stats.lognorm(0.9)
x4 = X4.rvs(1000, random_state=1)
x4g = np.linspace(0, 50, 1000)
plotDistribution(X4, x4, x4g)
```
