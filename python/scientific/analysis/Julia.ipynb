```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
%matplotlib inline
```

```python
mpl.rcParams['figure.figsize'] = 2.25*np.array(mpl.rcParams['figure.figsize'])
```

```python
phi = (1+np.sqrt(5))/2
```

```python
def julia(z, c=complex(-0.1, 0.65)):
    return z**2 + c
```

```python
def compute(func=julia, size=(1200, 1200), xlim=(-1.5, +1.5), ylim=(-1.5, +1.5), max_iter=2000, atol=10,
            cbar=True, grid=True, cmap='jet', norm=colors.LogNorm, return_axe=True,
            aspect='equal'):
    
    # Values:
    nit = np.zeros(size)
    for i in range(size[0]):
        for j in range(size[1]):
            z = complex(xlim[0] + i/(size[0]-1)*(xlim[1]-xlim[0]), ylim[0] + j/(size[1]-1)*(ylim[1]-ylim[0]))
            while (abs(z) <= atol) and (nit[i,j] < max_iter):
                z = func(z)
                nit[i,j] += 1
                
    # Figure:
    fig, axe = plt.subplots()
    pos = axe.imshow(nit.T, interpolation='nearest', aspect=aspect, origin='lower',
                     extent=list(xlim)+list(ylim), cmap=cmap, norm=norm(vmin=1, vmax=nit.max()))
    axe.set_xlabel(r"$\mathrm{Re}(z)$")
    axe.set_ylabel(r"$\mathrm{Im}(z)$")
    
    if grid:
        axe.grid()
    if cbar:
        cbar = fig.colorbar(pos)
        
    if return_axe:
        return nit, axe
    else:
        return nit
```

```python
x, axe = compute(return_axe=True)
axe.get_figure().savefig('Julia_01.png', dpi=120)
```

```python
x, axe = compute(xlim=(-0.25,0.25), ylim=(-0.25,0.25))
axe.get_figure().savefig('Julia_02.png', dpi=120)
```

```python
x, axe = compute(lambda x: julia(x, c=complex(1-phi, 0)))
axe.get_figure().savefig('Julia_03.png', dpi=120)
```
