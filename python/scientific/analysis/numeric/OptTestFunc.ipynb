```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm
from matplotlib import cm
%matplotlib inline
```

 - https://en.wikipedia.org/wiki/Himmelblau%27s_function
 - https://en.wikipedia.org/wiki/Test_functions_for_optimization

```python
#@np.vectorize
def Himmelblau(x):
    return (x[:,0]**2+x[:,1]-11)**2 + (x[:,0]+x[:,1]**2-7)**2
```

```python
x = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, x)
XX = np.array([X.ravel(), Y.ravel()]).T
```

```python
Z = Himmelblau(XX).reshape(X.shape)
```

```python
fig = plt.figure()
axe = fig.gca(projection='3d')
surf = axe.plot_surface(X, Y, Z, alpha=0.9, linewidth=0, antialiased=True,cmap=cm.jet)
axe.contour(X, Y, Z, np.logspace(-2, 3, 20, base=10), zdir='z', offset=0, cmap=cm.jet)
#axe.view_init(25, -30)
axe.set_title("Optimization Test Function: Himmelblau")
axe.set_xlabel(r"$x$")
axe.set_ylabel(r"$y$")
axe.set_zlabel(r"$z$")
fig.tight_layout()
fig.savefig("Himmelblau_Surface.png", dpi=120)
```

```python
fig, axe = plt.subplots()
CS = axe.contour(X, Y, Z, [10, 25, 50, 75, 100, 125, 150, 175, 180, 200, 250], linewidths=1, cmap=cm.jet)
axe.clabel(CS, inline=1, fontsize=8, fmt="%.0f")
axe.set_title("Optimization Test Function: Himmelblau")
axe.set_xlabel(r"$x$")
axe.set_ylabel(r"$y$")
axe.set_aspect('equal')
axe.grid()
fig.tight_layout()
fig.savefig("Himmelblau_Levels.png", dpi=120)
```
