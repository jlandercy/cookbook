````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib.tri import Triangulation
%matplotlib inline
````

### Moebus

````python
theta = np.linspace(0, 2 * np.pi, 50)
w = np.linspace(-0.25, 0.25, 10)
w, theta = np.meshgrid(w, theta)
phi = 0.5 * theta
````

````python
r = 1 + w * np.cos(phi)
x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z = np.ravel(w * np.sin(phi))
````

````python
tri = Triangulation(np.ravel(w), np.ravel(theta))
````

````python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='jet', antialiased=True)
#ax.set_xlim(-1, 1)
#ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1);
````

### Scatter

````python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
````

````python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)
````

````python
ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', linewidth=0.5)
````

````python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none', antialiased=True)#, rstride=1, cstride=1)
````

### Sphere

````python
theta = np.linspace(0, 2*np.pi, 20)
phi = np.linspace(0, np.pi, 20)
theta, phi = np.meshgrid(theta, phi)
rho = 1
````

````python
x = np.ravel(rho*np.cos(theta)*np.sin(phi))
y = np.ravel(rho*np.sin(theta)*np.sin(phi))
z = np.ravel(rho*np.cos(phi))
````

````python
tri = Triangulation(np.ravel(theta), np.ravel(phi))
````

````python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='jet', antialiased=True)
ax.get_figure().savefig('sphere.png', dpi=120)
#ax.set_xlim(-1, 1)
#ax.set_ylim(-1, 1)
#ax.set_zlim(-1, 1);
````

### Torus

````python
a, c = 1, 4
u = np.linspace(0, 2*np.pi, 20)
v = u.copy()
u, v = np.meshgrid(u, v)
````

````python
x = np.ravel((c + a*np.cos(v))*np.cos(u))
y = np.ravel((c + a*np.cos(v))*np.sin(u))
z = np.ravel(a*np.sin(v))
````

### Diabolo

- http://www.math.uri.edu/~bkaskosz/trans1/parsur1.html

```plot3d([cos(s)*cos(theta),cos(s)*sin(theta),s],theta=0..2*Pi,s=0..2*Pi);```

````python
# Parameters:
u = np.linspace(0, 1*np.pi, 20)
v = np.linspace(0, 2*np.pi, 20)
# Parametric Surface:
u, v = np.meshgrid(u, v)
x = np.ravel(np.cos(v)*np.cos(u))
y = np.ravel(np.cos(v)*np.sin(u))
z = np.ravel(v)
# Mesh Triangulation:
tri = Triangulation(np.ravel(u), np.ravel(v))
````

````python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='jet', antialiased=True)
ax.get_figure().savefig('torus.png', dpi=120)
````

### Mbius Strip

 - https://en.wikipedia.org/wiki/M%C3%B6bius_strip

````python
# Parameters:
u = np.linspace(0, 2*np.pi, 20)
v = np.linspace(-1, 1, 20)
# Parametric Surface:
u, v = np.meshgrid(u, v)
x = np.ravel((2 + (v/2)*np.cos(u/2))*np.cos(u))
y = np.ravel((2 + (v/2)*np.cos(u/2))*np.sin(u))
z = np.ravel(v/2*np.sin(u/2))
# Mesh Triangulation:
tri = Triangulation(np.ravel(u), np.ravel(v))
````

````python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='jet', antialiased=True)
ax.set_zlim([-1,1])
ax.get_figure().savefig('mobius.png', dpi=120)
````
