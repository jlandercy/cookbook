```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
class square_generator(stats.rv_continuous):
    
    def _pdf(self, x):
        return -6*(x - 1/2)*(x + 1/2)
    
    #def _cdf(self, x):
    #    return (1/2)*(-4*x**3 + 3*x + 1)
    
square = square_generator(a=-1/2, b=1/2, name="Square")
```

```python
def sample(loc=0.0, scale=1.0):
    X = square(loc=loc, scale=scale)
    data = X.rvs(size=1000)
    x = np.linspace(*(np.array(X.support()) + np.array([-0.5, 0.5])), 1000)
    print(X.support())
    fig, axe = plt.subplots()
    axe.hist(data, density=1, label="Data")
    axe.plot(x, X.pdf(x), label="PDF")
    axe.plot(x, X.cdf(x), label="CDF")
    axe.set_title("Square Distribution (loc={:.1f}, scale={:.1f})".format(loc, scale))
    axe.set_xlabel("Variable, $x$")
    axe.set_ylabel("Variable Distribution")
    axe.legend()
    axe.grid()
    return axe
```

```python
sample()
```

```python
sample(loc=1.)
```

```python
sample(scale=2)
```
