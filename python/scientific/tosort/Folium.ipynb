```python
import folium
import pandas as pd
import geopandas as gpd
```

```python
url1 = 'http://data-mobility.irisnet.be/resources/collectostops-2016-01-01.json?_ga=2.249737286.1755994789.1589030028-1714394335.1575117761'
url2 = 'http://data-mobility.irisnet.be/resources/pavedroads-2016-01-01.json?_ga=2.17407865.1755994789.1589030028-1714394335.1575117761'
url3 = 'https://mybrugis.irisnet.be/geoserver/ows?service=WFS&request=getFeature&typeNames=PER:Zone_voor_stedelijke_herwaardering_2020&outputFormat=application/json&_ga=2.249802822.1755994789.1589030028-1714394335.1575117761'
```

```python
df1 = gpd.read_file(url1)
df2 = gpd.read_file(url2)
df3 = gpd.read_file(url3)
```

https://python-visualization.github.io/folium/modules.html#folium.folium.Map

```python
m = folium.Map(location=[50.8503, 4.3517], width='70%', height='70%',
               tiles='OpenStreetMap', min_zoom=10, max_zoom=18, zoom_start=11,
               #maxBounds=[[50.7963, 4.3139], [50.9140, 4.4371]]
)
```

 - https://python-visualization.github.io/folium/modules.html#folium.features.GeoJson
 - https://leafletjs.com/reference-1.6.0.html#geojson

```python
def styler1(feature):
    return {
        #'icon': folium.Icon(icon='cloud'),
        'color': 'red'
    }
```

```python
fc1 = folium.GeoJson(df1, name='Collecto', style_function=styler1,
               tooltip=folium.features.GeoJsonTooltip(['nom', 'adres_fr', 'notes'])).add_to(m)
```

```python
def styler2(feature):
    return {
        'color': 'green'
    }
```

```python
fc2 = folium.GeoJson(df2, name='Paved Road').add_to(m)
```

```python
def styler3(feature):
    return {
        'color': 'red'
    }
```

```python
fc3 = folium.GeoJson(df3, style_function=styler3, name='Revitalization').add_to(m)
```

```python
lc = folium.LayerControl().add_to(m)
```
